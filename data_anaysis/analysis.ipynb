```python
import pandas
```

```python
csv = pandas.read_csv("wandb_export_2025-11-25T17_54_57.486-05_00.csv")
```

```python
csv
```

```python
import matplotlib.pyplot as plt

csv_processed = csv[~csv["complete_frozen_low_lr - episode_reward"].isna()]
csv_processed = csv_processed.rename(columns={
    "complete_frozen_low_lr - episode_reward": "frozen_resnet",
    "cnn_scratch - episode_reward": "cnn_scratch",
    "pca_resnet - episode_reward": "pca_resnet",
})

csv_processed["Step"] = (csv_processed["Step"] - 1024) // 1025

plt.plot(csv_processed["Step"], csv_processed["cnn_scratch"], label="Local Attn. CNN Encoder")
plt.plot(csv_processed["Step"], csv_processed["frozen_resnet"], label="Local Attn. ResNet18 Backbone")
plt.plot(csv_processed["Step"], csv_processed["pca_resnet"], label="Sliding Window Attn. with ResNet18 PCA")
plt.legend()
```
